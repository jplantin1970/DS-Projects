```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
csv_url ='https://raw.githubusercontent.com/MicrosoftDocs/mslearn-introduction-to-machine-learning/main/Data/ml-basics/diabetes.csv'
df = pd.read_csv(csv_url)
df.head()
```

```python
# Features distribution by Diabetics
features= ['Pregnancies',	'PlasmaGlucose',	'DiastolicBloodPressure',	'TricepsThickness',	'SerumInsulin',	'BMI',	'DiabetesPedigree','Age']

for col in features:
  df.boxplot(column=col, by= 'Diabetic', figsize=(8,5) )
  plt.title(col)
```

```python
# Load in sci-kit learn functions for constructing decision tree model
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

# Package to display the splitof the tree
from sklearn.tree import plot_tree

from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix
from sklearn.metrics import recall_score, precision_score, f1_score, accuracy_score
import sklearn.metrics as metrics
```

```python
# Separate Features and labels
X, y = df[['Pregnancies', 'PlasmaGlucose', 'DiastolicBloodPressure', 'TricepsThickness', 'SerumInsulin', 'BMI', 'DiabetesPedigree', 'Age']].values, df[['Diabetic']].values
print('Features:', X[:3], '\nLabels:', y[:3], sep= '\n')
```

```python
# Split dataset into training and testing
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.3, stratify=y, random_state=0)
print('Training cases: %d\nTest cases: %d' %(X_train.shape[0], X_test.shape[0]))
```

```python
# Instantiate the model
decision_tree=  DecisionTreeClassifier(random_state= 0)

# Fit the model to training data
clf= decision_tree.fit(X_train, y_train)

# Make predictions on test data
dt_pred = decision_tree.predict(X_test)
```

```python
# Generate performance metrics

print("Accuracy:", "%.3f" % accuracy_score(y_test, dt_pred))
print("Precision:", "%.3f" % precision_score(y_test, dt_pred))
print("Recall:", "%.3f" % recall_score(y_test, dt_pred))
print("F1 Score:", "%.3f" % f1_score(y_test, dt_pred))
```

```python
# Calculate the values for each quadrant in the confusion matrix
cm = metrics.confusion_matrix(y_test, dt_pred, labels=clf.classes_)
```

```python
# Generate the confusion matrix display

disp= metrics.ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=clf.classes_)
disp.plot()
```
